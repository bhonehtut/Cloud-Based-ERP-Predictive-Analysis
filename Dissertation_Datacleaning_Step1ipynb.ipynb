Load data set

```python
import pandas as pd

# Load the cleaned dataset
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv("/content/drive/MyDrive/Master Final Project/Dataset/erp_procurement_16000_clean.csv")

# ------------------------------------
# 1. Columns to drop
# ------------------------------------
drop_cols = [
    '_link','_link_awards','_link_main_x','_link_award','_link_main_y',
    '_link_tender','_link_bid_details','_link_main',
    'id','id_award','id_tender','tender_id','id_bid_details',
    'relatedLots','relatedLots_bid_details','relatedBid','relatedLot',
    'ocid','bids_url','buyer_id','tender_item_id','lot_id'
]

df = df.drop(columns=drop_cols, errors='ignore')

# ------------------------------------
# 2. Convert date columns to datetime
# ------------------------------------
date_cols = [
    'date_award','date_tender','tender_tenderPeriod_endDate',
    'metaData_modified','tender_document_datePublished'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# ------------------------------------
# 3. Create date features
# ------------------------------------
for col in date_cols:
    df[col + '_year'] = df[col].dt.year
    df[col + '_month'] = df[col].dt.month
    df[col + '_day'] = df[col].dt.day

# ------------------------------------
# 4. Remove original datetime columns
# ------------------------------------
df = df.drop(columns=date_cols)

# ------------------------------------
# 5. Save final modelling dataset
# ------------------------------------
df.to_csv("/content/drive/MyDrive/Master Final Project/Dataset/erp_procurement_model_ready.csv", index=False)

print("Final modelling dataset shape:", df.shape)
df.head()
```
